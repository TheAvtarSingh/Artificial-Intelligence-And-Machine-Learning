```python
# Program for demonstration of one hot encoding
  
# import libraries
import numpy as np
import pandas as pd
  
# import the data required
data = pd.read_csv("data.csv")
print(data.head())
```

```python
print(data['INDUS'].unique())
print(data['NOX'].unique())
```

```python
data['INDUS'].value_counts()
data['NOX'].value_counts()
```

```python
one_hot_encoded_data = pd.get_dummies(data, columns = ['INDUS', 'NOX'])
print(one_hot_encoded_data)
```
