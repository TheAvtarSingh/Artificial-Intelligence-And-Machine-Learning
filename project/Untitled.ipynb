```python
# label Encoding on diamonds
import pandas as pd
df = pd.read_csv('diamonds.csv')
df.head()
```

```python

# Ordinal Labeling


cut_labels = {'Ideal':0, 
               'Fair':1, 
               'Good':2,
               'Very Good':3,
               'Premium':4}
df['cut'] = df['cut'].map(cut_labels)
df
```

```python
# Label Encoing

df['cut'] = le.fit_transform(df['cut'])
df['color'] = le.fit_transform(df['color'])

df
```

```python
# One hot encoding
dummies = pd.get_dummies(df['color'])
dummies
```

```python
# adding into data
df = pd.concat([df, dummies], axis = 1)
df
```

```python
# Removing previous color

df.drop(['color'], axis = 1, inplace = True)
df
```
