```python
data = pd.read_csv('new.csv')
```

```python
x = data.x.to_numpy()
y = data.y.to_numpy()
```

```python
pp.scatter(x,y)
```

```python
x = x/500.0
y = y/500.0 
# diving thr numbers by 500 each
```

```python
m,c = np.random.random(),np.random.random()
def learn(x,y,m,c,epoch):
 for i in range(epoch):
    error = y-( m * x + c )
    x_error = ( y- ( m * x + c ) ) * x #element wise
    error = np.sum ( error ) / len ( x )
    x_error = np.sum ( x_error ) / len ( x )
    learning_rate = 0.01
    delta_m = learning_rate * x_error
    delta_c =  learning_rate* error
    m += delta_m
    c += delta_c
 return (m,c)
```

```python
m,c = learn(x,y,m,c,2000)
```

```python
x1 = [0,1]
y1 = [0*m+c,1*m+c]
```

```python
pp.scatter(x,y)
pp.plot(x1,y1)
```
